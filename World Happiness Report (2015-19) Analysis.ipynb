These countries are islands and territories that haven't been listed in our continent dictionary. As the number of missing values is small, and there are duplicates of country values, we can fill them manually. 

```python
#Handle the missing values in the Region column 

df.loc[df['Country'] == 'Somaliland region', 'Country'] = 'Somaliland Region'

df.loc[df['Country'] == 'Puerto Rico', 'Region'] = 'North America'
df.loc[df['Country'] == 'North Cyprus', 'Region'] = 'Europe'
df.loc[df['Country'] == 'Somaliland Region', 'Region'] = 'Africa'
df.loc[df['Country'] == 'Hong Kong S.A.R., China', 'Region'] = 'Asia'
df.loc[df['Country'] == 'Taiwan Province of China', 'Region'] = 'Asia'
df.loc[df['Country'] == 'Macedonia', 'Region'] = 'Europe'
```

```python
#show the records that have missing values in the Trust column

df.loc[df['Trust (Government Corruption)'].isnull() == True]
```

Because United Arab Emirates is in Asia, we'll fill its missing values with the mean values of those columns for Asian countries, instead of the mean of Trust and Dystopia Residual of the whole dataset.

```python
#Handle the missing values for United Arab Emirates in 2018

asia_avg_trust = df.loc[df['Region'] == 'Asia'].groupby('Region')['Trust (Government Corruption)'].mean()

asia_avg_dystopia = df.loc[df['Region'] == 'Asia'].groupby('Region')['Dystopia Residual'].mean()


df['Trust (Government Corruption)'].replace([np.nan], asia_avg_trust, inplace=True)

df['Dystopia Residual'].replace([np.nan], asia_avg_dystopia, inplace=True)
```

There are some outliers in the Family, Trust (i.e., absence of government corruption), Generosity and Dystopia Residual columns. However, the number of outliers is small, and the magnitude of the outliers in the distribution is not significant. 

From the graph, we can tell the Dystopia Residual column has the most outliers. Therefore, we could investigate this column further.

```python
#Investigating the outliers in the Generosity distribution

#Calculate the 25th and 75th percentiles of the 
q1 = df['Dystopia Residual'].quantile(0.25)
q3 = df['Dystopia Residual'].quantile(0.75)

#Find the IQR
IQR = q3 - q1
factor = 1.5

#Calculate the lower limit 
lower_limit = q1 - IQR * factor

#Calculate the upper limit
upper_limit = q3 + IQR * factor

#Using outlier limits for filtering

#Create a mask for values lower than lower_limit
is_lower = df['Dystopia Residual'] < lower_limit

#Create a mask for values higher than upper_limit
is_higher = df['Dystopia Residual'] > upper_limit

#Combine the masks to filter for outliers
outliers = df['Dystopia Residual'][is_lower | is_higher]

#Count and print the number of outliers
print(len(outliers))
```

The Dystopia Residual column has 19 outliers. 

# Observations

- From 2015 to 2019, all of the happiest countries in the world were in Europe, while all of the least happy countries were in Africa. 
- Interestingly, two Asian countries, Uzbekistan and Cambodia, ranked as the top countries in the world for freedom, surpassing Norway, Denmark, and Finland.
- Happiness Score is highly correlated with Economic Production (indicated by GDP per Capita), Family (or Social Support), and Health (indicated by Life Expectancy). Among the six factors surveyed, these three appear to have the biggest impact on the happiness of a nation, much more than the other three. Therefore, a policymaker might want to put an emphasis on these 3 factors. 
- Oceania is the happiest continent in the world, with a significantly higher average happiness score than other continents. Europe, South America, and North America have relatively similar levels of happiness. Africa is the least happy region. 
- The world's happiness plunged from 2016 to 2017, but then increased significantly from 2017 to 2019.
